```python
df = pd.read_csv("./data/201902-fordgobike-tripdata.csv")
df.head()
```

## Process categorical variables

```python
cat_cols = df.select_dtypes(include = ['object']).columns.to_list()
```

```python
def label_encoder(df, cat_cols):
    """
    This function takes in a dataframe and a list of categorical columns and returns a dataframe with the categorical columns encoded.

    Args:
        df (dataframe): dataframe to be encoded
        cat_cols (list): list of categorical columns

    Returns:
        dataframe: dataframe with categorical columns encoded
    """
    for i in cat_cols:
        le = LabelEncoder()
        df[i] = le.fit_transform(df[i])
    return df
```

```python
df = label_encoder(df, cat_cols)
df.head()
```

## Handle missing values

```python
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
# Total missing value
df.isna().sum()
```

Enhance dataset with new date related fields

```python
# Convert type of start_time and end_time to datetime
df['start_time']=pd.to_datetime(df['start_time'])
df['end_time']=pd.to_datetime(df['end_time'])
df.info(verbose=True)
```

```python
# Add a field start time month name
df['start_time_month_name']=df['start_time'].dt.strftime('%B')
# Add a field start time month number
df['start_time_month']=df['start_time'].dt.month.astype(int)
# Add a field start time weekdays
df['start_time_weekday']=df['start_time'].dt.strftime('%a')
# Add a field start time day
df['start_time_day']=df['start_time'].dt.day.astype(int)
# Add a field start time hour
df['start_time_hour']=df['start_time'].dt.hour
```

```python
# Log transform duration_sec
df['log_duration_sec'] = np.log10(df['duration_sec'])
```

```python
# Calculate age from member_birth_year (assuming current year is 2024)
current_year = 2023
df['member_age'] = current_year - df['member_birth_year']
```

```python
df = df[df['member_age']<=60]
```

```python
#Generate a new field for member age group from member_age_bin
df['member_age_bins'] = df['member_age'].apply(lambda x: '10 - 20' if 10<x<=20
                                                  else '20 - 30' if 20<x<=30
                                                  else '30 - 40' if 30<x<=40
                                                  else '40 - 50' if 40<x<=50
                                                  else '50 - 60' if 50<x<=60
                                                  else x)
```

```python
def set_plot_labels(xlabel, ylabel, title):
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
```

## Analysis, Modeling, Visualization

### Question 1: What is the distribution of trip durations in the dataset?

Calculations provided serve the purpose of visually and numerically summarizing the distribution of trip durations, specifically after applying a logarithmic transformation

```python
# Assuming duration_sec is in a list or DataFrame
durations = df['log_duration_sec'].tolist()

# Basic statistics
mean_duration = np.mean(durations)
median_duration = np.median(durations)
min_duration = np.min(durations)
max_duration = np.max(durations)

# Plotting histogram
plt.figure(figsize=(8, 6))
sb.histplot(durations, bins=20, kde=True)
set_plot_labels('Duration (seconds)', 'Count', 'Distribution of Trip Durations')
plt.show()
```

Answer: The histogram shows that trip durations are right-skewed, with a peak towards shorter durations and a long tail extending towards longer durations.
This range helps in understanding the variability of trip durations and identifying any exceptionally short or long trips that may warrant further investigation.

### Question 2: What is the distribution of user types in the GoBike sharing system?

```python
# Calculate counts of each user type
user_type_counts = df['user_type'].value_counts()

# Plotting pie chart
plt.figure(figsize=(8, 6))
plt.pie(user_type_counts, labels=user_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('User Type Split for GoBike Sharing System')
plt.legend(['0: Customer', '1: Subcriber'])
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Answer: According to the chart, we see that the number of subscribers is more than customers, nearly 10 times more

### Question 3: How does the percentage of bike rides vary across different age groups from 20 to 60 years old?

```python
trip_by_age = df.groupby('member_age_bins').agg({'bike_id':'count'})
```

```python
trip_by_age['percent'] = (trip_by_age['bike_id']/trip_by_age['bike_id'].sum())*100
```

```python
trip_by_age['percent'].plot(kind='bar', figsize=(12,8))
set_plot_labels('Age group', 'Pecentage(%)', 'Percentage of all bike rides per age group from 20 to 60 years old')
plt.xticks(rotation=360)
```

Answer: The plot helps identify which age groups contribute the most to bike ride usage within the specified range of 20 to 60 years old.
The bar plot visually compares the percentage of bike rides across different age groups

## Correlation

```python
# Heatmap: Correlation matrix
correlation_matrix = df[['duration_sec', 'member_age', 'start_station_latitude']].corr()
plt.figure(figsize=(8, 6))
sb.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

The heatmap shows a weak positive correlation between duration_sec and member_age, with a correlation coefficient of around 0.003
The heatmap reveals no significant correlation between duration_sec and start_station_latitude. The correlation coefficient is close to zero, indicating that the latitude of the starting station does not influence ride duration.
